```python
import pandas as pd
```

```python
battery_csv = "./original_datasets/battery_2021.csv"

latlng_csv = "./original_datasets/2021_counties_california.csv"
```

```python
battery_df = pd.read_csv(battery_csv)
latlng_df = pd.read_csv(latlng_csv)
```

```python
battery_df.head()
```

```python
latlng_df.head()
```

```python
merged_df = pd.merge(battery_df, latlng_df, on="COUNTY")
merged_df.head()
```

```python
batteries_df = merged_df.drop(columns=['Public Level 1', 'Shared Private Level 1', 'Public Level 2', 'Shared Private Level 2', 'Public DC Fast', 'Shared Private DC Fast', 'STATE'])
batteries_df.head()
```

```python
batteries_df['Technology'] = 'battery'
```

```python
batteries_df['Open Retail Status'] = 'Open Retail'
```

```python
batteries_df.head()
```

```python
updatedBAT_df = batteries_df.rename(columns={"LAT": "Latitude", "LNG": "Longitude"})
updatedBAT_df.head()
```

```python
hydrogen_csv = "./original_datasets/hydrogen_2021.csv"
```

```python
hydrogen_df = pd.read_csv(hydrogen_csv)
hydrogen_df.head()
```

```python
hydrogen_df['Technology'] = 'hydrogen'
hydrogen_df.head()
```

```python
updatedH2_df = hydrogen_df[['Open Retail Status', 'COUNTY', 'Latitude', 'Longitude', 'Fueling Positions', 'Technology']]
```

```python
updatedH2_df.head()
```

```python
updatedH2_df = updatedH2_df.rename(columns={"Fueling Positions":"Total"})
```

```python
updatedH2_df.head()
```

```python
ZEV_df = pd.concat([updatedBAT_df, updatedH2_df])
```

```python
ZEV_df
```

```python
ZEV_df.to_csv("ZEV_CA_infrastructure.csv", index=True)
```
